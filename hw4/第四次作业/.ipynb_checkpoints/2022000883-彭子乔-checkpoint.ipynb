# 1.对两个数据集进行描述性统计分析

## 1.1航空公司客户信息数据

```python
cus_info = pd.read_csv("./air_data.csv",encoding='gb18030')
cus_info.head()
```

```python
cus_info.describe().T
```

```python
print(cus_info.info())
```

```python
# 查看是否含有空值
cus_info.isnull().sum().sort_values(ascending=False).head()
```

```python
# 删除票价为空的数据
cus_info = cus_info.dropna(subset = ['SUM_YR_1','SUM_YR_2'])
```

```python
# 删除两年票价为0的数据
sum1 = cus_info['SUM_YR_1'] != 0
sum2 = cus_info['SUM_YR_2'] != 0
cus_info = cus_info[sum1 | sum2]
```

```python
#整理出数值型特征和类别型特征
all_cols = cus_info.columns.tolist()
numerical_cols = []
categorical_cols = []

for col in all_cols:
    if (cus_info[col].dtype != 'object') :
        numerical_cols.append(col)
    else:
        categorical_cols.append(col)

print('数值型变量数目为：',len(numerical_cols))
print('类别型变量数目为：',len(categorical_cols))
```

```python
#对连续性数值变量绘制箱型图筛选异常值
fig = plt.figure(figsize=(40,30),dpi=120)
for i in range(len(numerical_cols)):
    _ = plt.subplot(6, 6, i+1)
    _ = sns.boxplot(y=cus_info[numerical_cols[i]],orient='v', width=0.5)
    _ = plt.ylabel(numerical_cols[i], fontsize=20)
plt.show()
```

## 1.2航空公司客户满意度调查数据

```python
ps = pd.read_csv("./Airline_Passenger_Satisfaction.csv")
ps.head(5)
```

```python
# 剔除无用数据
ps = ps.drop('Unnamed: 0', axis=1)
ps = ps.drop('id', axis=1)
```

```python
ps.isnull().sum()
```

```python
#填充缺失值
ps["Arrival Delay in Minutes"] = ps["Arrival Delay in Minutes"].fillna(ps["Arrival Delay in Minutes"].mean())
```

```python
#整理出数值型特征和类别型特征
all_cols = ps.columns.tolist()
numerical_cols = []
categorical_cols = []

for col in all_cols:
    if (ps[col].dtype != 'object') :
        numerical_cols.append(col)
    else:
        categorical_cols.append(col)

print('数值型变量数目为：',len(numerical_cols))
print('类别型变量数目为：',len(categorical_cols))
```

```python
for col in categorical_cols:
    _ = plt.figure(figsize=(12,2))
    _ = sns.countplot(x=col,data=ps,palette=["#2DBA6E", "#DC2C66", "#E5C26A"])
    _ = plt.show()
```

```python
for col in numerical_cols:
    _ = plt.figure(figsize=(12,2))
    _ = sns.histplot(ps[col])
    plt.show()
```

### 通过以上分析我们可以得出以下结论：
- 不满意的乘客比满意的多；
- 忠诚客户比不忠诚客户多；
- 商务客户比个人客户多；
- 年龄主要在20-60岁；
- 飞行里程主要在1000以下。

```python
# 画出相关性热力图
_ = plt.figure(figsize=(20,12))
_ = sns.heatmap(ps[numerical_cols].corr(), annot=True, cmap="coolwarm", annot_kws={"fontsize":12}, fmt=".2f")
plt.show()
```

```python
# "Departure Delay in Minutes"和"Arrival Delay in Minutes"强相关，可以删除其中一个
ps = ps.drop('Arrival Delay in Minutes', axis=1)
```

# 2.对Airline_Passenger_Satisfaction.csv的特征进行主成分分析和因子分析，及其对“航空公司满意度水平”的影响,并解释分析结果。

```python
# 对文本列进行数字编码
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ps['Gender'] = le.fit_transform(ps['Gender'])
ps['Customer Type'] = le.fit_transform(ps['Customer Type'])
ps['Type of Travel'] = le.fit_transform(ps['Type of Travel'])
ps['Class'] = le.fit_transform(ps['Class'])
ps['satisfaction'] = le.fit_transform(ps['satisfaction'])
```

```python
# 特征值进行缩放
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(ps)
```

```python
scaled_data = scaler.transform(ps)
scaled_data
```

```python
#只取6-20行的评分作为特征进行分析
x = ps[ps.columns[6:20]]
```

## 2.1主成分分析

```python
# 进行PCA降维
from sklearn.decomposition import PCA
ev,v = np.linalg.eig(x.cov())
ev
ev_sorted = sorted(ev,reverse=True)
# 可视化
# plot横轴是指标个数，纵轴是ev值
# scatter横轴是指标个数，纵轴是ev值

_ = plt.scatter(range(1,x.shape[1]+1),ev_sorted)
_ = plt.plot(range(1,x.shape[1]+1),ev_sorted)
_ = plt.title('Screen Plot')
_ = plt.xlabel('Components')
_ = plt.ylabel('Eigenvalue')
plt.show()
```

```python
pca = PCA(n_components=2)
pca.fit(x)
```

```python
x_pca = pca.transform(x)
```

```python
scaled_data.shape
x_pca.shape
```

```python
plt.figure(figsize=(12,6))
_ = plt.scatter(x_pca[:,0],x_pca[:,1],c=ps["satisfaction"],cmap='rainbow')
_ = plt.xlabel('First principal component')
_ = plt.ylabel('Second Principal Component')
```

```python
pca.components_
df=pd.DataFrame(pca.transform(x))#输出降维后的特征可用于之后的分析
df
```

```python
map= pd.DataFrame(pca.components_,columns=x.columns)
plt.figure(figsize=(12,6))
sns.heatmap(map)
```

### 结果分析：
- 在评分中食品和饮料、座椅舒适度、机上娱乐和清洁度得分更高更容易让乘客满意；
- 出发和达到时间方便以及是否易于在网上订购对乘客满意度影响不大；
- 所以航空公司应更多地提升食品和饮料、座椅舒适度、机上娱乐和清洁度等方面。

# 3.至少使用两种聚类方法对air_data.csv进行聚类分析，并比较各方法的异同。

### 原始数据集特征较多，选取8个重要特征:第一年总票价、第二年总票价、观测窗口总飞行公里数、飞行次数、平均乘机时间间隔、入会时间、观测窗口的结束时间、平均折扣率。

```python
cus_info = cus_info[[ "FFP_DATE", "LOAD_TIME", "FLIGHT_COUNT", "SUM_YR_1", "SUM_YR_2", "SEG_KM_SUM", "AVG_INTERVAL" , "avg_discount"]]
cus_info[0:5]
```

```python
cus_info["LOAD_TIME"] = pd.to_datetime(cus_info["LOAD_TIME"])
cus_info["FFP_DATE"] = pd.to_datetime(cus_info["FFP_DATE"])
cus_info["TIME"] = cus_info["LOAD_TIME"] - cus_info["FFP_DATE"]
cus_info["TIME"] = cus_info["TIME"].astype(np.int64)/(60*60*24*10**9)
cus_info["AVG_FARE"] = (cus_info["SUM_YR_1"] + cus_info["SUM_YR_2"]) / cus_info["SEG_KM_SUM"]
cus_info = cus_info[["TIME", "FLIGHT_COUNT", "AVG_FARE", "SEG_KM_SUM", "AVG_INTERVAL", "avg_discount"]]
cus_info.head(5)
```

```python
cus_info = (cus_info - cus_info.mean(axis=0))/(cus_info.std(axis=0))
```

## 3.1K-means聚类

```python
sse = []
for each in range(1,30):
    km = KMeans(n_clusters=each,random_state=0)
    km.fit(cus_info)
    sse.append(km.inertia_)
ax = pd.Series(sse).plot(kind='line',figsize=(12,6),marker='o',grid=True,xticks=range(0,29))   #xticks：在哪些位置显示x轴刻度标签
_ = ax.set(xticklabels=range(1,30))     
```

```python
# 构建KMeans模型训练数据
cluster = KMeans(n_clusters=5,random_state=123).fit(cus_info)
# 获取聚类结果
y_pred = cluster.labels_
# 获取质心
centers = cluster.cluster_centers_
# 查看簇内平方和
inertia = cluster.inertia_ 
```

```python
#统计不同类别样本的数目
r1 = pd.Series(cluster.labels_).value_counts().sort_index()
print('最终每个类别的数目为：\n',r1)
```

```python
cluster_data = pd.DataFrame(cluster.cluster_centers_, 
                            columns= cus_info.columns)

result = pd.concat([pd.DataFrame(r1, columns=['聚类个数'])
                    ,cluster_data], axis=1)
result
```
